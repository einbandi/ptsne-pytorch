```python
import numpy as np
import pandas as pd
```

```python
data = np.loadtxt('projected_output_test.csv')
```

```python
data = data.reshape(15, 60000, 2)
```

```python
data_sub = data[:,::60]
```

```python
data_sub = data_sub.transpose(1,0,2).reshape(-1,2)
```

```python
from torchvision.datasets import MNIST
labels = MNIST(
        './data',
        train=True,
        download=True
    ).targets
```

```python
labels_sub = np.repeat(labels[::60],15)
```

```python
from matplotlib import pyplot as plt
```

```python
plt.scatter(data_sub[:,0],data_sub[:,1], c=labels_sub)
```

```python
epochs = np.tile(np.arange(15),1000)
```

```python
line = np.repeat(np.arange(1000),15)
```

```python
altered_labels = labels_sub
altered_labels[altered_labels == 8] = 0
altered_labels[altered_labels == 9] = 0
```

```python
df = pd.DataFrame()
df['x'] = data_sub[:,0]
df['y'] = data_sub[:,1]
df['line'] = line
df['algo'] = altered_labels
```
